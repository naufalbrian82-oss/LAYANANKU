```python
print("=====input nilai=====")

nilai = float(input("Masukkan nilai Anda: "))

if nilai >= 90:
    print("Grade: A (Lulus)")
elif nilai >= 80:
    print("Grade: B (Lulus)")
elif nilai >= 75:
    print("Grade: C (Lulus)")
else:
    print("Grade: D (Tidak lulus)")
```

```python
print("=====mengitung berat badan ideal=====")

tinggi = float(input("masukan tinggi badan (cm)"))
berat = float(input("berat badan (kg)"))

tinggi_m= tinggi/100
bmi = berat / (tinggi_m ** 2)
print(f"\nBMI Anda: {bmi:.2f}")

if bmi < 18.5:
    print("Kategori: Kurus")
elif 18.5 <= bmi < 25:
    print("Kategori: Normal (Ideal)")
elif 25 <= bmi < 30:
    print("Kategori: Berat berlebih")
else:
    print("Kategori: Obesitas")
```

```python
while True:
    print("\n=== PROGRAM MENU PILIHAN ===")
    print("1. Penjumlahan")
    print("2. Hitung Luas Segitiga")
    print("3. Keluar")

    pilihan = input("Pilih menu (1/2/3): ")

    if pilihan == "1":
        print("\n--- Penjumlahan ---")
        a = float(input("Masukkan angka pertama: "))
        b = float(input("Masukkan angka kedua: "))
        print("Hasil =", a + b)

    elif pilihan == "2":
        print("\n--- Hitung Luas Segitiga ---")
        alas = float(input("Masukkan alas: "))
        tinggi = float(input("Masukkan tinggi: "))
        luas = 0.5 * alas * tinggi
        print("Luas segitiga =", luas)

    elif pilihan == "3":
        print("\nProgram selesai. Terima kasih!")
        break

    else:
        print("Pilihan tidak valid, coba lagi!")
```

```python
import math

while True:
    print("\n=== PROGRAM BANGUN DATAR ===")
    print("1. Luas Persegi")
    print("2. Luas Lingkaran")
    print("3. Luas Segitiga")
    print("4. Keluar")

    pilihan = input("Pilih menu (1/2/3/4): ")

    if pilihan == "1":
        print("\n--- Luas Persegi ---")
        sisi = float(input("Masukkan panjang sisi: "))
        luas = sisi * sisi
        print("Luas Persegi =", luas)

    elif pilihan == "2":
        print("\n--- Luas Lingkaran ---")
        r = float(input("Masukkan jari-jari: "))
        luas = math.pi * r * r
        print("Luas Lingkaran =", luas)

    elif pilihan == "3":
        print("\n--- Luas Segitiga ---")
        alas = float(input("Masukkan alas: "))
        tinggi = float(input("Masukkan tinggi: "))
        luas = 0.5 * alas * tinggi
        print("Luas Segitiga =", luas)

    elif pilihan == "4":
        print("\nProgram selesai.")
        break

    else:
        print("Pilihan tidak valid!")
```

```python
while True:
    print("\n=== KALKULATOR SEDERHANA ===")
    print("1. Penjumlahan")
    print("2. Pengurangan")
    print("3. Perkalian")
    print("4. Pembagian")
    print("5. Keluar")

    pilihan = input("Pilih menu (1-5): ")

    if pilihan in ["1", "2", "3", "4"]:
        a = float(input("Masukkan angka pertama: "))
        b = float(input("Masukkan angka kedua: "))

        if pilihan == "1":
            print("Hasil =", a + b)
        elif pilihan == "2":
            print("Hasil =", a - b)
        elif pilihan == "3":
            print("Hasil =", a * b)
        elif pilihan == "4":
            if b != 0:
                print("Hasil =", a / b)
            else:
                print("Error: Pembagi tidak boleh 0!")

    elif pilihan == "5":
        print("Program selesai.")
        break

    else:
        print("Pilihan tidak valid!")
```
